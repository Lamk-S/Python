# Pandas Arithmetic Operations & Data Alignment

En este cuaderno exploraremos cómo realizar operaciones matemáticas con estructuras de datos de Pandas (`Series` y `DataFrames`). A diferencia de las listas tradicionales de Python, Pandas está optimizado para operaciones vectorizadas y maneja automáticamente la alineación de índices, lo cual es vital al trabajar con datos incompletos del mundo real.

```python
import pandas as pd

# Escenario: Ventas de dos sucursales diferentes durante la semana
branch_a_sales = pd.Series([150, 200, 350, 400, 550], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
branch_b_sales = pd.Series([50, 150, 250, 350, 450], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

print("--- Basic Series Arithmetic ---")
print("Total Sales (A + B):\n", branch_a_sales + branch_b_sales)
print("\nDifference (A - B):\n", branch_a_sales - branch_b_sales)
```

## Operations with DataFrames
Las operaciones aritméticas en DataFrames se aplican elemento por elemento de manera vectorizada, basándose en la alineación de índices y columnas.

```python
# Escenario: Inventario inicial y artículos vendidos en dos tiendas
initial_inventory = pd.DataFrame({
    'Laptops': [15, 25, 35],
    'Monitors': [21, 31, 41],
}, index=['Store_1', 'Store_2', 'Store_3'])

items_sold = pd.DataFrame({
    'Laptops': [2, 8, 12],
    'Monitors': [13, 25, 22],
}, index=['Store_1', 'Store_2', 'Store_3'])

print("--- DataFrame Arithmetic ---")
remaining_inventory = initial_inventory - items_sold
print("Remaining Inventory:\n", remaining_inventory)
```

## Broadcasting (Operations with Constants)
El *Broadcasting* permite aplicar una operación matemática entre un arreglo (como una Serie o DataFrame) y un valor escalar (constante). Pandas propaga este valor a todos los elementos automáticamente.

```python
print("--- Broadcasting ---")
tax_rate = 1.18 # Impuesto del 18%

# Aplicando el impuesto a todas las ventas de la Sucursal A
sales_with_tax = branch_a_sales * tax_rate
print("Branch A Sales with 18% Tax:\n", sales_with_tax)
```

## Explicit Arithmetic Methods & Data Alignment
Cuando utilizamos operadores nativos (`+`, `-`, `*`, `/`) con índices que no coinciden perfectamente, Pandas introduce valores nulos (`NaN`). 

Para evitar la pérdida de datos en análisis productivos, es mejor usar los métodos explícitos (`.add()`, `.sub()`, `.mul()`, `.div()`) junto con el parámetro `fill_value`.

```python
# Escenario: Ingresos de dos líneas de productos que se lanzaron en meses distintos
product_x_revenue = pd.Series([1000, 2000, 3000], index=['Jan', 'Feb', 'Mar'])
product_y_revenue = pd.Series([1500, 2500, 3500], index=['Feb', 'Mar', 'Apr'])

print("--- The Problem with Native Operators (+) ---")
# Enero y Abril devolverán NaN porque no existen en ambas series
print(product_x_revenue + product_y_revenue) 

print("\n--- The Solution: .add() with fill_value ---")
# Asume 0 si el mes falta en alguna de las series
total_revenue = product_x_revenue.add(product_y_revenue, fill_value=0)
print(total_revenue)
```
