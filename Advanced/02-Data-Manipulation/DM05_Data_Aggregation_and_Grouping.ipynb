# Data Aggregation and Grouping

La agregación de datos es una de las técnicas más poderosas en el análisis de datos. Nos permite segmentar nuestro conjunto de datos en grupos y aplicar cálculos matemáticos (como sumas, promedios o conteos) a cada segmento. En este cuaderno exploraremos el uso del método `.groupby()` y cómo realizar análisis de tendencias.

```python
import pandas as pd

# Escenario 1: Transacciones simples de productos por categoría
data = {
    "Categoria": ["A", "B", "A", "C", "B", "A"],
    "Valor": [10, 20, 30, 40, 50, 60],
    "Cantidad": [4, 1, 6, 2, 3, 5]
}

df = pd.DataFrame(data)
print("--- Dataset de Transacciones ---")
df
```

## 1. Agrupaciones Simples (Single Grouping)
El método `.groupby()` divide los datos basándose en los valores de una columna. Luego, encadenamos una función estadística para obtener un resultado resumido.

```python
print("--- Suma de Valores por Categoría ---")
suma_por_categoria = df.groupby("Categoria")["Valor"].sum()
print(suma_por_categoria)

print("\n--- Promedio de Valores por Categoría ---")
media_por_categoria = df.groupby("Categoria")["Valor"].mean()
print(media_por_categoria)

print("\n--- Frecuencia (Conteo) por Categoría ---")
recuento_por_categoria = df.groupby("Categoria")["Valor"].count()
print(recuento_por_categoria)
```

## 2. Múltiples Funciones con `.agg()`
En lugar de calcular métricas una por una, el método `.agg()` nos permite aplicar múltiples funciones estadísticas simultáneamente a nuestros grupos, devolviendo un DataFrame muy limpio.

```python
print("--- Múltiples Agregaciones Simultáneas ---")
agregados_por_categoria = df.groupby("Categoria")["Valor"].agg(["sum", "mean", "count"])
agregados_por_categoria
```

## 3. Agrupación Multinivel (MultiIndex)
Podemos agrupar por más de una categoría pasando una lista al método `.groupby()`. Esto crea una jerarquía o "MultiIndex".

```python
# Añadimos una nueva dimensión a nuestros datos (ej. Sucursal o Grupo)
df["Grupo"] = ["G1", "G1", "G2", "G2", "G1", "G2"]

print("--- Suma de Valores por Grupo y Categoría ---")
suma_por_grupo_categoria = df.groupby(["Grupo", "Categoria"])["Valor"].sum()
suma_por_grupo_categoria
```

## 4. Ejercicio Práctico: Análisis de Tendencias de Ventas
En el mundo real, la agrupación se utiliza a menudo junto con datos temporales para analizar el rendimiento a lo largo del tiempo.

```python
# Escenario 2: Registro de ventas diarias
datos_ventas = {
    "Fecha": ["2026-10-01", "2026-10-01", "2026-10-02", "2026-10-02", "2026-10-03", "2026-10-03"],
    "Producto": ["Producto A", "Producto B", "Producto A", "Producto B", "Producto A", "Producto B"],
    "Categoria": ["Electronica", "Ropa", "Electronica", "Ropa", "Electronica", "Ropa"],
    "Ventas": [100, 150, 200, 100, 300, 250]
}

df_ventas = pd.DataFrame(datos_ventas)
# Convertimos la columna 'Fecha' al tipo datetime para un manejo correcto de series de tiempo
df_ventas["Fecha"] = pd.to_datetime(df_ventas["Fecha"])

print("--- Ventas Totales por Categoría ---")
ventas_por_categoria = df_ventas.groupby("Categoria")["Ventas"].sum()
print(ventas_por_categoria)

print("\n--- Tendencia de Ventas Diarias ---")
ventas_por_fecha = df_ventas.groupby("Fecha")["Ventas"].sum()
print(ventas_por_fecha)

print("\n--- Desglose Diario por Producto ---")
ventas_por_fecha_producto = df_ventas.groupby(["Fecha", "Producto"])["Ventas"].sum()
print(ventas_por_fecha_producto)
```
