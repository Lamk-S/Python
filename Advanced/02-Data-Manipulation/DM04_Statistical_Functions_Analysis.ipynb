# Statistical Functions & Exploratory Data Analysis (EDA)

Las funciones estadísticas son fundamentales para entender la distribución, tendencia y variabilidad de nuestros datos. En este cuaderno exploraremos las operaciones matemáticas integradas que nos permiten realizar un Análisis Exploratorio de Datos (EDA) rápido y efectivo.

```python
import pandas as pd
import numpy as np

# Escenario: Métricas mensuales de operaciones y ventas de una tienda
# Hemos incluido intencionalmente algunos valores nulos (np.nan) para demostrar el conteo
data = {
    'Units_Sold': [120, 150, np.nan, 200, 250],
    'Revenue_USD': [1200, 1500, 1800, 2000, 2500],
    'Customer_Rating': [4.2, 4.5, 4.0, np.nan, 4.8]
}

df = pd.DataFrame(data, index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])

print("--- Monthly Store Operations Dataset ---")
df
```

## 1. Measures of Central Tendency
Estas funciones nos ayudan a encontrar el centro o el valor típico de nuestros datos.
- **`.mean()`**: Calcula el promedio aritmético. Puede ser sensible a valores atípicos (outliers).
- **`.median()`**: Encuentra el valor central. Es más robusto cuando hay valores atípicos extremos.

```python
print("--- Mean (Average) ---")
# Promedio de una columna específica
avg_units = df['Units_Sold'].mean()
print(f"Average Units Sold: {avg_units}")

print("\n--- Median (Central Value) ---")
# Mediana de todo el DataFrame
median_metrics = df.median()
print(median_metrics)
```

## 2. Measures of Dispersion
Nos indican qué tan "esparcidos" o alejados están los datos respecto a su media.
- **`.std()` (Standard Deviation)**: Muestra la desviación promedio de los datos. Valores altos indican mayor variabilidad.
- **`.var()` (Variance)**: Es el cuadrado de la desviación estándar.

```python
print("--- Standard Deviation ---")
std_dev = df.std()
print(std_dev)

print("\n--- Variance ---")
variance = df.var()
print(variance)
```

## 3. Extremes and Data Completeness
- **`.min()` / `.max()`**: Identifican los valores más bajos y más altos de nuestras series.
- **`.count()`**: Es vital para la limpieza de datos, ya que **solo cuenta los valores no nulos**. Nos ayuda a identificar rápidamente si faltan datos en alguna columna.

```python
print("--- Maximum Values ---")
max_values = df.max()
print(max_values)

print("\n--- Minimum Values ---")
min_values = df.min()
print(min_values)

print("\n--- Count (Non-null values) ---")
# Nota cómo 'Units_Sold' y 'Customer_Rating' mostrarán 4 en lugar de 5
data_completeness = df.count()
print(data_completeness)
```
