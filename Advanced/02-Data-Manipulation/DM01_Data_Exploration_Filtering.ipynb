# 📊 Exploración y Filtrado de Datos Reales

En este módulo daremos un paso más allá. Trabajaremos con un conjunto de datos real: **Denuncias de violencia contra la mujer**. Este dataset contiene información geográfica (UBIGEO, Provincias, Departamentos de Perú) y temporal.

Aprenderemos a inspeccionar la información, navegar por sus índices y aplicar filtros lógicos para extraer conclusiones significativas.

```python
import pandas as pd

# Cargamos los datos usando punto y coma como delimitador
df_vio = pd.read_csv('DM01_ViolenciaDenunciaMujer.csv', delimiter=';')

# Observamos los nombres de las columnas para entender la estructura
print("Columnas disponibles:", df_vio.columns.tolist())

# Mostramos las primeras 5 filas para verificar la carga correcta
df_vio.head()
```

## Navegando por Índices con `loc`

El método `.loc[]` nos permite acceder a un grupo de filas y columnas por sus etiquetas o una matriz booleana. Es fundamental para extraer registros específicos.

```python
# 1. Extraer una sola fila por su índice (retorna una Serie)
print("--- Fila 13 ---")
print(df_vio.loc[13])

# 2. Extraer múltiples filas específicas
df_vio.loc[[10, 13, 15, 18, 21]]

# 3. Extraer filas específicas y mostrar solo ciertas columnas
df_vio.loc[[10, 13, 15, 18, 21], ['AÑO', 'CANTIDAD']]
```

## Filtrado Condicional y Datos Faltantes

En el análisis de datos, es vital identificar registros con valores nulos (vacíos) y poder filtrar la información usando operadores lógicos (`>`, `<`, `==`).

```python
# Buscar filas donde el año no fue registrado (Valores NaN)
df_nulos = df_vio[df_vio['AÑO'].isna()]
print(f"Registros sin año: {len(df_nulos)}")

# Filtrar denuncias masivas (más de 100 casos en un mes/distrito)
df_vio[df_vio['CANTIDAD'] > 100]

# Filtrar por un departamento en específico (ej. LA LIBERTAD)
df_vio.loc[df_vio['DPTO_HECHO'] == 'LA LIBERTAD'].head(20)
```

## Análisis de Categorías

Para responder preguntas de negocio o investigación, Pandas nos ofrece herramientas poderosas para contar y agrupar categorías de forma instantánea.

```python
# 1. unique(): Ver todos los valores distintos de una columna
departamentos = df_vio['DPTO_HECHO'].unique()
print("Departamentos registrados:", departamentos)

# 2. value_counts(): Cuenta la frecuencia de cada valor (Ordenado de mayor a menor por defecto)
# Muy útil para ver qué departamento tiene más registros en el dataset
conteo_dptos = df_vio['DPTO_HECHO'].value_counts()
print("\nTop 5 departamentos con más registros:")
print(conteo_dptos.head())

# 3. isin(): Filtra usando una lista de valores permitidos.
# Ideal para evitar hacer múltiples condiciones con "==" y el operador "or" (|).
regiones_sur = ['AREQUIPA', 'MOQUEGUA', 'TACNA', 'PUNO', 'CUSCO']
df_sur = df_vio[df_vio['DPTO_HECHO'].isin(regiones_sur)]

df_sur.head()
```
