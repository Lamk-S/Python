# DVP02: Histograms and Data Distribution

A diferencia de un gráfico de barras (que compara categorías distintas), un **Histograma** se utiliza para analizar una sola variable numérica continua. Nos agrupa los datos en "cubetas" (bins) para mostrarnos cómo se distribuyen las frecuencias.

En este ejercicio, analizaremos las calificaciones de un examen universitario utilizando la clásica escala vigesimal (0 a 20).

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generamos un conjunto de 100 calificaciones simulando una "Campana de Gauss" (Distribución Normal)
# Media (loc) de 14, Desviación Estándar (scale) de 3.
calificaciones = np.random.normal(loc=14, scale=3, size=100)

# 2. Limitamos (clip) los valores para que no superen el rango real de notas (0 a 20)
calificaciones = np.clip(calificaciones, 0, 20)

# 3. Crear el Histograma
plt.figure(figsize=(10, 6))

# 'bins=10' divide el rango de 0-20 en 10 intervalos iguales (ej. 0-2, 2-4, 4-6...)
plt.hist(
    calificaciones, 
    bins=10, 
    color='#5bc0de', 
    edgecolor='black', # El borde negro es vital en los histogramas para distinguir los bins
    alpha=0.8
)

# 4. Estilizar el gráfico
plt.title("Distribución de Calificaciones del Examen Final", fontsize=15, pad=15)
plt.xlabel("Rango de Calificaciones (0 - 20)", fontsize=12)
plt.ylabel("Número de Estudiantes (Frecuencia)", fontsize=12)

# Añadimos una línea vertical roja para marcar la media aritmética
plt.axvline(calificaciones.mean(), color='red', linestyle='dashed', linewidth=2, label=f'Media: {calificaciones.mean():.1f}')
plt.legend()

plt.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()
```
