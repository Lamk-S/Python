# DIE02: Advanced CSV Configuration Options

Los archivos CSV (Comma-Separated Values) rara vez vienen en un formato perfecto. En el mundo real, los delimitadores cambian (especialmente en regiones que usan la coma para decimales), las codificaciones fallan y los tipos de datos se interpretan incorrectamente. 

En este cuaderno aprenderemos a dominar la función `.read_csv()` mediante sus parámetros de configuración clave.

## Key Parameters for Data Parsing
- **`sep`** o **`delimiter`**: Define el separador (ej. `,`, `;`, `\t` para tabulaciones).
- **`header`**: Indica qué fila usar como nombre de columnas (ej. `header=None`).
- **`names`**: Permite asignar nombres manuales a las columnas.
- **`encoding`**: Fundamental para leer caracteres especiales como la 'ñ' o tildes (ej. `utf-8`, `latin1`).
- **`nrows`**: Lee solo una porción del archivo, útil para archivos gigantes.
- **`na_values`**: Define qué strings deben ser considerados como valores nulos (`NaN`).
- **`dtype`**: Fuerza un tipo de dato específico para ciertas columnas, ahorrando memoria.

```python
import pandas as pd

# Primero, creamos el archivo CSV localmente para nuestro ejercicio práctico
csv_content = """marca;modelo;año;color;combustible;transmision;puertas;precio_usd;kilometraje
Toyota;Corolla;2022;Blanco;Gasolina;Automática;4;22500;18000
Honda;Civic;2021;Negro;Gasolina;Manual;4;21000;25000
Ford;Focus;2020;Azul;Diésel;Manual;4;17500;40000
Chevrolet;Cruze;2019;Gris;Gasolina;Automática;4;16000;52000
Nissan;Sentra;2023;Rojo;Gasolina;CVT;4;24500;8000
Hyundai;Elantra;2022;Plata;Híbrido;Automática;4;23800;15000
Kia;Forte;2021;Blanco;Gasolina;Automática;4;19800;27000
Volkswagen;Golf;2020;Azul;Diésel;Manual;5;20500;36000
Mazda;Mazda3;2023;Negro;Gasolina;Automática;4;26500;6000
Subaru;Impreza;2022;Verde;Gasolina;CVT;5;25500;12000"""

with open("DIE02_cars.csv", "w", encoding="utf-8") as f:
    f.write(csv_content)

print("Archivo 'DIE02_cars.csv' generado exitosamente.")
```

```python
# Leer el archivo CSV aplicando múltiples configuraciones simultáneamente:
df_cars = pd.read_csv(
    'DIE02_cars.csv',
    sep=';',                    # El delimitador es punto y coma, no coma
    encoding='utf-8',           # Asegura la correcta lectura de caracteres
    nrows=8,                    # Leemos solo las primeras 8 filas para exploración inicial
    dtype={
        'año': 'int64', 
        'puertas': 'int64', 
        'precio_usd': 'float64', 
        'kilometraje': 'int64'
    }                           # Optimizamos y aseguramos los tipos de datos matemáticos
)

print("--- DataFrame from Configured CSV ---")
df_cars
```
