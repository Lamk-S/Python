# DIE01: JSON Data Handling (Import & Export)

El formato **JSON** (JavaScript Object Notation) es el estándar universal para el intercambio de datos en la web. Ya sea que estés consumiendo una API Rest, extrayendo datos de una página web (scraping) o leyendo archivos de configuración, saber manejar JSON es una habilidad crítica en el análisis de datos.

En este cuaderno aprenderemos a:
1. Exportar diccionarios de Python a archivos JSON.
2. Leer archivos JSON locales hacia DataFrames de Pandas.
3. Consumir datos JSON directamente desde una URL externa.

```python
import json
import pandas as pd

# Escenario: Catálogo de especificaciones de hardware
# Representado como una lista de diccionarios (estructuralmente idéntico a un JSON array)
laptops = [
    { "name": "Dell XPS 13", "price": 1299.99, "brand": "Dell", "processor": "Intel Core Ultra 7" },
    { "name": "MacBook Air 15", "price": 1499.00, "brand": "Apple", "processor": "Apple M3" }, 
    { "name": "HP Spectre x360 14", "price": 1399.99, "brand": "HP", "processor": "Intel Core Ultra 7" },
    { "name": "ASUS ROG Zephyrus G14", "price": 1799.99, "brand": "ASUS", "processor": "AMD Ryzen 9 8945HS" },
    { "name": "MacBook Pro 16", "price": 3499.00, "brand": "Apple", "processor": "Apple M4 Max" }
]

print("Datos originales en memoria (Python List of Dictionaries):")
for laptop in laptops[:2]: # Imprimimos solo los dos primeros para visualizar
    print(laptop)
```

## 1. Writing Data to a JSON File
Antes de usar Pandas, es común interactuar con la librería nativa de Python `json` para volcar (`dump`) nuestros datos a un archivo físico. Esto es muy útil cuando hemos recolectado o limpiado datos y necesitamos guardarlos para compartirlos.

```python
# Generar el archivo físico en formato JSON
file_path = "DIE01_laptops.json"

with open(file_path, "w", encoding="utf-8") as f:
    json.dump(laptops, f, indent=4) # indent=4 lo hace legible para humanos (Pretty Print)

print(f"Archivo exportado exitosamente como: {file_path}")
```

## 2. Reading Local JSON Files with Pandas
Una vez que tenemos el archivo físico, Pandas puede leerlo e interpretar su estructura jerárquica directamente a una tabla bidimensional (DataFrame) usando el método `.read_json()`.

```python
# Cargar el archivo JSON físico al DataFrame
df_laptops = pd.read_json(file_path)

print("--- DataFrame from Local JSON ---")
df_laptops
```

## 3. Reading Remote JSON Data (URLs / APIs)
No siempre necesitamos descargar el archivo localmente. Si los datos están alojados en un servidor o un repositorio público (como GitHub), Pandas puede acceder a ellos directamente pasándole la URL.

```python
# Cargar un dataset clásico (Iris) alojado en un repositorio remoto
url_iris = "https://raw.githubusercontent.com/domoritz/maps/master/data/iris.json"
df_iris = pd.read_json(url_iris)

print("--- DataFrame from Remote URL ---")
df_iris.head() # Mostramos solo las primeras 5 filas para explorar la estructura
```
