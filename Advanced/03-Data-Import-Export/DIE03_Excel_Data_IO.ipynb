# DIE03: Excel Data I/O (Input/Output)

Excel sigue siendo la herramienta de negocio más utilizada en el mundo corporativo. Como analistas de datos, necesitamos poder extraer datos de hojas de cálculo complejas y, del mismo modo, exportar nuestros resultados a formatos `.xlsx` para que los clientes o gerentes puedan consumirlos.

Para interactuar con archivos Excel modernos, Python utiliza el motor `openpyxl`.

```python
# Instalación del motor necesario para leer/escribir archivos .xlsx modernos
# En entornos de Jupyter, usamos '!' y '%' para ejecutar comandos de terminal
!python -m pip install --upgrade pip
%pip install openpyxl
```

```python
import pandas as pd

# Escenario: Catálogo de inventario de una tienda de electrónica
inventario_data = [
    {"Producto": "Laptop",     "Marca": "Dell",      "Precio": 1200, "Stock": 15},
    {"Producto": "Smartphone", "Marca": "Samsung",   "Precio": 850,  "Stock": 30},
    {"Producto": "Tablet",     "Marca": "Apple",     "Precio": 600,  "Stock": 20},
    {"Producto": "Monitor",    "Marca": "LG",        "Precio": 320,  "Stock": 12},
    {"Producto": "Teclado",    "Marca": "Logitech",  "Precio": 45,   "Stock": 50},
    {"Producto": "Mouse",      "Marca": "Razer",     "Precio": 30,   "Stock": 65},
    {"Producto": "Audífonos",  "Marca": "Sony",      "Precio": 150,  "Stock": 25},
    {"Producto": "Impresora",  "Marca": "HP",        "Precio": 250,  "Stock": 10}
]

df_inventario = pd.DataFrame(inventario_data)

# Guardar el DataFrame en un archivo Excel
# CRÍTICO: index=False evita que se exporte la columna numérica del índice (0, 1, 2...)
df_inventario.to_excel('DIE03_electronics.xlsx', sheet_name='Productos', index=False)

print("Archivo Excel generado exitosamente.")
df_inventario
```

```python
# Leer el archivo Excel especificando explícitamente el motor
df_productos_importados = pd.read_excel('DIE03_electronics.xlsx', engine='openpyxl')

print("--- DataFrame importado desde Excel ---")
df_productos_importados
```
