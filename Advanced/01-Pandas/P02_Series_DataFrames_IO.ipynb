# 🧬 Estructuras de Datos y Lectura de Archivos en Pandas

Antes de analizar grandes volúmenes de datos, necesitamos entender los bloques de construcción de Pandas. Existen dos estructuras fundamentales:
1. **Series:** Arreglos unidimensionales (como una sola lista o una columna).
2. **DataFrames:** Estructuras bidimensionales (como una tabla o una hoja de cálculo).

Además, Pandas nos permite cargar datos fácilmente desde archivos externos.

```python
# Si no tienes pandas instalado en tu entorno virtual, descomenta la siguiente línea:
# !pip install pandas

import pandas as pd
```

## 1. Trabajando con Series

Una Serie es más que una simple lista de Python; tiene un "índice" que nos permite etiquetar cada valor. Veamos un ejemplo representando los IVs (Valores Individuales) de un Pokémon destinado al competitivo.

```python
# Datos base
ivs_data = [31, 31, 31, 29, 31, 31]
stats_labels = ['HP', 'Ataque', 'Defensa', 'Atq. Esp.', 'Def. Esp.', 'Velocidad']

# Creamos la Serie asignando las etiquetas personalizadas al índice
serie_ivs = pd.Series(data=ivs_data, index=stats_labels, name="IVs Garchomp")

# Mostramos la serie
serie_ivs
```

## 2. Creando DataFrames desde Diccionarios

Un DataFrame es esencialmente una colección de Series que comparten el mismo índice. Podemos estructurar información más compleja, por ejemplo, para llevar un registro de crianza competitiva.

```python
# Cada clave del diccionario se convierte en el nombre de una columna
crianza_dict = {
    'Pokémon': ['Bagon', 'Larvitar', 'Gible', 'Beldum'],
    'Naturaleza': ['Alegre', 'Firme', 'Alegre', 'Firme'],
    'Habilidad': ['Cabeza Roca', 'Agallas', 'Piel Tosca', 'Cuerpo Puro'],
    'EVs_Entrenados': [True, False, True, False]
}

# Convertimos a DataFrame
df_crianza = pd.DataFrame(crianza_dict)

# Mostramos la tabla
df_crianza
```

## 3. Importando datos desde un CSV

En el mundo real, los datos rara vez se escriben a mano en el código. Lo más común es importarlos desde bases de datos o archivos CSV (*Comma Separated Values*).

```python
# Utilizamos pd.read_csv para cargar el archivo.
# El parámetro 'delimiter' es crucial cuando nuestro archivo usa punto y coma (;) en lugar de comas (,).
df_csv = pd.read_csv('P02_Data.csv', delimiter=';')

# head() muestra solo las primeras 5 filas. 
# Es una buena práctica para inspeccionar los datos sin saturar la memoria ni la pantalla.
df_csv.head()
```
