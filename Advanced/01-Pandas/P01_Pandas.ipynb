# 🐼 Introducción a Pandas: El poder de los DataFrames

**Pandas** es la biblioteca estándar de Python para la manipulación y análisis de datos. Si alguna vez has trabajado con hojas de cálculo o bases de datos relacionales, te sentirás como en casa.

En este primer ejercicio, aprenderemos a crear la estructura principal de Pandas: el **DataFrame**.

```python
import pandas as pd

# 1. Definimos nuestros datos base usando un diccionario estándar de Python
# Las claves representan los nombres de las columnas, y las listas son los datos.
data = {
    'Nombre': ['Juan', 'María', 'Pedro', 'Ana'],
    'Edad': [28, 34, 22, 19],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla'],
    'Salario': [30000, 40000, 25000, 35000]
}

# 2. Convertimos el diccionario en un DataFrame bidimensional
df = pd.DataFrame(data)

# 3. Al colocar la variable al final, Jupyter la renderiza como una tabla HTML formateada
df
```

## Entendiendo la estructura

Al ejecutar la celda anterior, notarás que Pandas añadió automáticamente una columna a la izquierda con números (0, 1, 2, 3). A esto se le conoce como el **Índice (Index)**, y sirve como el identificador único de cada fila.

## Exploración rápida de los datos

Cuando trabajamos con miles de registros, imprimir todo el DataFrame no es viable. Pandas nos ofrece métodos muy potentes para obtener resúmenes instantáneos de la información.

```python
# Muestra un resumen técnico: tipos de datos, valores nulos y uso de memoria
print("--- INFORMACIÓN DEL DATAFRAME ---")
df.info()

print("\n--- RESUMEN ESTADÍSTICO ---")
# describe() calcula automáticamente promedios, mínimos, máximos y percentiles de las columnas numéricas
df.describe()
```

## Aplicando Filtros (El verdadero poder)

Tener los datos en una tabla está bien, pero Pandas brilla cuando necesitamos extraer información específica mediante filtros (conocido como *Boolean Indexing*).

```python
# Ejemplo: ¿Quiénes ganan más de 25,000 y son menores de 30 años?
filtro = (df['Salario'] > 25000) & (df['Edad'] < 30)

# Aplicamos el filtro al DataFrame original
resultados = df[filtro]

resultados
```
